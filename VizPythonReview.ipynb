```python
x = np.linspace (-3,3, 100)
print(x)
plt.plot(x,x, label = 'y = x')
plt.plot(x, x**2, label = 'y = x squared')
plt.plot(x, x**3, label = 'y = x cubed')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from importlib import reload
plt=reload(plt)

x = np.linspace (-3,3, 100)
print(x)
plt.plot(x,x)
plt.plot(x, x**2)
plt.plot(x, x**3)
plt.legend(['y = x',  'y = x squared', 'y = x cubed'])
plt.xlabel('x')
plt.ylabel ('f(x)')
plt.xlim(x[0], x[-1]) # extens lines to limits of the box to show that 
        #function is contiguos 
plt.ylim (x[0], x[-1]) # better to hardcode to -10, 10
plt.title('Plot of functions')
# plt.axis('square')
# plt.gca().set_aspect('auto')
# plt.gca().set_aspect('equal')
# plt.gca().set_aspect(0.9)
plt.gca().set_aspect(1/plt.gca().get_data_ratio()) # To get a square plot precisely
plt.show()
```

```python
import matplotlib.pyplot as plt
from importlib import reload
plt=reload(plt)

# Doing same things with subplots instead of plt.

fig, ax = plt.subplots()
ax.plot(x,x)
ax.plot(x, x**2)
ax.plot(x, x**3)
ax.legend(['y = x',  'y = x squared', 'y = x cubed'])
ax.set_xlabel('x')
ax.set_ylabel ('f(x)')
ax.set_title ('This is a plot of three functions')
ax.set_xlim(x[0], x[-1]) # extens lines to limits of the box to show that 
        #function is contiguos 
ax.set_ylim (x[0], x[-1]) # better to hardcode to -10, 10
# plt.axis('square')
# plt.gca().set_aspect('auto')
# plt.gca().set_aspect('equal')
# plt.gca().set_aspect(0.9)
ax.set_aspect (1/ ax.get_data_ratio())
ax.grid()
```
