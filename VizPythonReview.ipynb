```python
# Hilbert Matrix

N = 1000

H = np.zeros ((N, N))

for i in range (1, N + 1) :
  for j in range (1, N + 1) :
    H [i - 1,j -1 ] = 1 / (i + j - 1)

import seaborn as sns
sns.heatmap (H, vmin = 0, vmax = 0.002)
plt.show()
```
